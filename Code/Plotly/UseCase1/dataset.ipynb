```python
from datetime import datetime, timedelta

def safe_extract(data, indices):
    return [data[i] for i in indices] if data else []


def subtract_days(data):
    for days, details in data.items():
        try:
            days_delta = timedelta(days=float(days))  # Create a timedelta object for the given days
        except ValueError:
            # Skip non-numeric keys
            continue
        updated_times = []
        for dt_string in details['dt_time']:
            dt_obj = datetime.fromisoformat(dt_string)  # Parse the datetime string to a datetime object
            new_dt_obj = dt_obj - days_delta  # Subtract the days
            new_dt_string = new_dt_obj.isoformat()  # Convert back to ISO format string
            new_dt_string = new_dt_string.replace('T', ' ')  # Replace 'T' with a space
            updated_times.append(new_dt_string)  # Append the formatted datetime string to the list
        details['dt_time'] = updated_times  # Update the dictionary with the new times

    return data

def divide_by_model(data):
    divided_data = {}
    for key in data:
        divided_data[key] = {}
        model_indices = {}
        
        # Extract unique models and their indices
        models = data[key]['model']
        if models != []:
            for idx, model in enumerate(models):
                if model not in model_indices:
                    model_indices[model] = []
                model_indices[model].append(idx)
            
            for model, indices in model_indices.items():
                divided_data[key][model] = {
                    'dt_time': [data[key]['dt_time'][i] for i in indices],
                    'model': [data[key]['model'][i] for i in indices],
                    'AlongTrack': {
                        'dt': safe_extract(data[key]['AlongTrack']['dt'], indices),
                        'upper_bound': safe_extract(data[key]['AlongTrack']['upper_bound'], indices),
                        'lower_bound': safe_extract(data[key]['AlongTrack']['lower_bound'], indices),
                    },
                    'CrossTrack': {
                        'dt': safe_extract(data[key]['CrossTrack']['dt'], indices),
                        'upper_bound': safe_extract(data[key]['CrossTrack']['upper_bound'], indices),
                        'lower_bound': safe_extract(data[key]['CrossTrack']['lower_bound'], indices),
                    },
                    'Radial': {
                        'dt': safe_extract(data[key]['Radial']['dt'], indices),
                        'upper_bound': safe_extract(data[key]['Radial']['upper_bound'], indices),
                        'lower_bound': safe_extract(data[key]['Radial']['lower_bound'], indices),
                    }
                }
            divided_data[key]
        else:
            divided_data[key][0.0] = data[key]
    return divided_data
```

```python
def calculate_mean(df, column_name):
    group = df.groupby('creation_date')[column_name].mean().tolist()

    if np.isnan(group).all():
        group = []
    
    return group

def add(df, type, str_type, value):
    df[type]['Forecast'][str_type]['AlongTrack']['dt'] += value['Forecast'][str_type]['AlongTrack']['dt']
    df[type]['Forecast'][str_type]['AlongTrack']['upper_bound'] += value['Forecast'][str_type]['AlongTrack']['upper_bound']
    df[type]['Forecast'][str_type]['AlongTrack']['lower_bound'] += value['Forecast'][str_type]['AlongTrack']['lower_bound']
    df[type]['Forecast'][str_type]['CrossTrack']['dt'] += value['Forecast'][str_type]['CrossTrack']['dt']
    df[type]['Forecast'][str_type]['CrossTrack']['upper_bound'] += value['Forecast'][str_type]['CrossTrack']['upper_bound']
    df[type]['Forecast'][str_type]['CrossTrack']['lower_bound'] += value['Forecast'][str_type]['CrossTrack']['lower_bound']
    df[type]['Forecast'][str_type]['Radial']['dt'] += value['Forecast'][str_type]['Radial']['dt']
    df[type]['Forecast'][str_type]['Radial']['upper_bound'] += value['Forecast'][str_type]['Radial']['upper_bound']
    df[type]['Forecast'][str_type]['Radial']['lower_bound'] += value['Forecast'][str_type]['Radial']['lower_bound']

    return df

def zip_data(column_name, value):

    paired_data_target = list(zip(value['Forecast']['Target']['dt_time'], value['Forecast']['Target'][column_name]['dt'], value['Forecast']['Target'][column_name]['upper_bound'], value['Forecast']['Target'][column_name]['lower_bound']))
    paired_data_chaser = list(zip(value['Forecast']['Chaser']['dt_time'], value['Forecast']['Chaser'][column_name]['dt'], value['Forecast']['Chaser'][column_name]['upper_bound'], value['Forecast']['Chaser'][column_name]['lower_bound']))

    sorted_target = sorted(paired_data_target, key=lambda x: datetime.fromisoformat(x[0]))
    sorted_chaser = sorted(paired_data_chaser, key=lambda x: datetime.fromisoformat(x[0]))

    if len(sorted_target) > 0:
        target_sorted_dt_time, target_sorted_dt, target_sorted_upper_bound, target_sorted_lower_bound = zip(*sorted_target)

        value['Forecast']['Target']['dt_time'] = list(target_sorted_dt_time)
        value['Forecast']['Target']['dt'] = list(target_sorted_dt)
        value['Forecast']['Target']['upper_bound'] = list(target_sorted_upper_bound)
        value['Forecast']['Target']['lower_bound'] = list(target_sorted_lower_bound)

    if len(sorted_chaser) > 0:
        chaser_sorted_dt_time, chaser_sorted_dt, chaser_sorted_upper_bound, chaser_sorted_lower_bound = zip(*sorted_chaser)

        value['Forecast']['Chaser']['upper_bound'] = list(chaser_sorted_upper_bound)
        value['Forecast']['Chaser']['lower_bound'] = list(chaser_sorted_lower_bound)
        value['Forecast']['Chaser']['dt_time'] = list(chaser_sorted_dt_time)
        value['Forecast']['Chaser']['dt'] = list(chaser_sorted_dt)
    return value

def make_dataset(df): 
    tca = df['tca'][0]
    
    tag = df['tags'].unique()

    datasets = {}

    for t in tag:
        split1 = t.split(',')
        split = split1[1].split('_')
        dt_time = list(df[df['tags'] == t]['creation_date'])
        if 'target' and 'chaser' in split:
            n = split[0] + ' vs ' + split[0]
            target = split[0]
            chaser = split[0]
            color = '#FFD521'
        elif 'target' in split:
            n = split[0] + ' vs HAC'
            target = split[0]
            chaser = 'HAC'
            color = '#00ACC1'
        elif 'chaser' in split:
            n = 'HAC vs ' + split[0]
            target = 'HAC'
            chaser = split[0]
            color = '#DDEA1F'
        elif 'baseline' in split:
            n = 'HAC vs HAC'
            target = 'HAC'
            chaser = 'HAC'
            color = '#A1173C'
        else:
            print('Error')

        table = df[df['tags'] == t]
        target_forecast = table[table['is_target_pred'] == True]
        chaser_forecast = table[table['is_chaser_pred'] == True]

        datasets[n] = {
            'tca': tca, 'name': n, 'target': target, 'chaser': chaser, 'dt_time': dt_time, 'color': color,
            'dt': {
                'Target':{
                    'AlongTrack': list(table['target_along_track']), 
                    'CrossTrack': list(table['target_cross_track']), 
                    'Radial': list(table['target_radial']),
                },
                'Chaser':{
                    'AlongTrack': list(table['chaser_along_track']), 
                    'CrossTrack': list(table['chaser_cross_track']), 
                    'Radial': list(table['chaser_radial']),
                },
            },
            'Forecast':{
                'Target':{
                    'dt_time': target_forecast['creation_date'].tolist(),
                    'model': target_forecast['prediction_time_to_tca'].tolist(),
                    'AlongTrack': {
                        'dt' : target_forecast['along_track_pred'].tolist(),
                        'upper_bound': target_forecast['along_track_upper_bound'].tolist(),
                        'lower_bound': target_forecast['along_track_lower_bound'].tolist(),
                    
                    }, 
                    'CrossTrack': {
                        'dt': target_forecast['cross_track_pred'].tolist(),
                        'upper_bound': target_forecast['cross_track_upper_bound'].tolist(),
                        'lower_bound': target_forecast['cross_track_lower_bound'].tolist(),
                    
                    }, 
                    'Radial': {
                        'dt': target_forecast['radial_pred'].tolist(),
                        'upper_bound': target_forecast['radial_upper_bound'].tolist(),
                        'lower_bound': target_forecast['radial_lower_bound'].tolist(),
                    }

                },
                'Chaser':{
                    'dt_time': chaser_forecast['creation_date'].tolist(),
                    'model': chaser_forecast['prediction_time_to_tca'].tolist(),
                    'AlongTrack': {
                        'dt': chaser_forecast['along_track_pred'].tolist(),
                        'upper_bound': chaser_forecast['along_track_upper_bound'].tolist(),
                        'lower_bound': chaser_forecast['along_track_lower_bound'].tolist(),
                    
                    }, 
                    'CrossTrack': {
                        'dt': chaser_forecast['cross_track_pred'].tolist(),
                        'upper_bound': chaser_forecast['cross_track_upper_bound'].tolist(),
                        'lower_bound': chaser_forecast['cross_track_lower_bound'].tolist(),
                    
                    }, 
                    'Radial': {
                        'dt': chaser_forecast['radial_pred'].tolist(),
                        'upper_bound': chaser_forecast['radial_upper_bound'].tolist(),
                        'lower_bound': chaser_forecast['radial_lower_bound'].tolist(),
                    },
                }
            }
        }

    new_datasets = {}

    for value in datasets.values():
        if value['chaser'] not in new_datasets:
            new_datasets[value['chaser']] = {
                'name': value['chaser'],
                'tca': value['tca'],
                'Target': {
                    'dt_time': [],
                    'AlongTrack': [],
                    'CrossTrack': [],
                    'Radial': [],
                },
                'Chaser': {
                    'dt_time': [],
                    'AlongTrack': [],
                    'CrossTrack': [],
                    'Radial': [],
                },
                'Forecast':{
                    'Target':{
                        'dt_time': [],
                        'model': [],
                        'AlongTrack': {
                            'dt' : [],
                            'upper_bound': [],
                            'lower_bound': [],
                        
                        }, 
                        'CrossTrack': {
                            'dt': [],
                            'upper_bound': [],
                            'lower_bound': [],

                        
                        }, 
                        'Radial': {
                            'dt': [],
                            'upper_bound': [],
                            'lower_bound': [],
                        }

                    },
                    'Chaser':{
                        'dt_time': [],
                        'model': [],
                        'AlongTrack': {
                            'dt': [],
                            'upper_bound': [],
                            'lower_bound': [],
                        
                        }, 
                        'CrossTrack': {
                            'dt': [],
                            'upper_bound': [],
                            'lower_bound': [],
                        
                        }, 
                        'Radial': {
                            'dt': [],
                            'upper_bound': [],
                            'lower_bound': [],
                        },
                    }
                } 
            }
        if value['target'] not in new_datasets:
            new_datasets[value['target']] = {
                'name': value['target'],
                'tca': value['tca'],
                'Target': {
                    'dt_time': [],
                    'AlongTrack': [],
                    'CrossTrack': [],
                    'Radial': [],
                },
                'Chaser': {
                    'dt_time': [],
                    'AlongTrack': [],
                    'CrossTrack': [],
                    'Radial': [],
                },
                'Forecast':{
                    'Target':{
                        'dt_time': [],
                        'model': [],
                        'AlongTrack': {
                            'dt' : [],
                            'upper_bound': [],
                            'lower_bound': [],
                        
                        }, 
                        'CrossTrack': {
                            'dt': [],
                            'upper_bound': [],
                            'lower_bound': [],

                        
                        }, 
                        'Radial': {
                            'dt': [],
                            'upper_bound': [],
                            'lower_bound': [],
                        }

                    },
                    'Chaser':{
                        'dt_time': [],
                        'model': [],
                        'AlongTrack': {
                            'dt': [],
                            'upper_bound': [],
                            'lower_bound': [],
                        
                        }, 
                        'CrossTrack': {
                            'dt': [],
                            'upper_bound': [],
                            'lower_bound': [],
                        
                        }, 
                        'Radial': {
                            'dt': [],
                            'upper_bound': [],
                            'lower_bound': [],
                        },
                    }
                }
            }
        
        target = value['target']
        new_datasets[target]['Target']['dt_time'] += value['dt_time']
        new_datasets[target]['Target']['AlongTrack'] += value['dt']['Target']['AlongTrack']
        new_datasets[target]['Target']['CrossTrack'] += value['dt']['Target']['CrossTrack']
        new_datasets[target]['Target']['Radial'] += value['dt']['Target']['Radial']
        new_datasets[target]['Forecast']['Target']['dt_time'] += value['Forecast']['Target']['dt_time']
        new_datasets[target]['Forecast']['Target']['model'] += value['Forecast']['Target']['model']
        new_datasets = add(new_datasets, target, 'Target', value)
    
        chaser = value['chaser']
        new_datasets[chaser]['Chaser']['dt_time'] += value['dt_time']
        new_datasets[chaser]['Chaser']['AlongTrack'] += value['dt']['Chaser']['AlongTrack']
        new_datasets[chaser]['Chaser']['CrossTrack'] += value['dt']['Chaser']['CrossTrack']
        new_datasets[chaser]['Chaser']['Radial'] += value['dt']['Chaser']['Radial']
        new_datasets[chaser]['Forecast']['Chaser']['dt_time'] += value['Forecast']['Chaser']['dt_time']
        new_datasets[chaser]['Forecast']['Chaser']['model'] += value['Forecast']['Chaser']['model']
        new_datasets = add(new_datasets, chaser, 'Chaser', value)

    for value in new_datasets.values():
        paired_data_target = list(zip(value['Target']['dt_time'], value['Target']['AlongTrack'], value['Target']['CrossTrack'], value['Target']['Radial']))
        paired_data_chaser = list(zip(value['Chaser']['dt_time'], value['Chaser']['AlongTrack'], value['Chaser']['CrossTrack'], value['Chaser']['Radial']))

        sorted_target = sorted(paired_data_target, key=lambda x: datetime.fromisoformat(x[0]))
        sorted_chaser = sorted(paired_data_chaser, key=lambda x: datetime.fromisoformat(x[0]))

        if len(sorted_target) > 0:
            target_sorted_dt_time, target_sorted_AlongTrack, target_sorted_CrossTrack, target_sorted_Radial = zip(*sorted_target)
        
            value['Target']['dt_time'] = list(target_sorted_dt_time)
            value['Target']['AlongTrack'] = list(target_sorted_AlongTrack)
            value['Target']['CrossTrack'] = list(target_sorted_CrossTrack)
            value['Target']['Radial'] = list(target_sorted_Radial)

        if len(sorted_chaser) > 0:
            chaser_sorted_dt_time, chaser_sorted_AlongTrack, chaser_sorted_CrossTrack, chaser_sorted_Radial = zip(*sorted_chaser)

            value['Chaser']['dt_time'] = list(chaser_sorted_dt_time)
            value['Chaser']['AlongTrack'] = list(chaser_sorted_AlongTrack)
            value['Chaser']['CrossTrack'] = list(chaser_sorted_CrossTrack)
            value['Chaser']['Radial'] = list(chaser_sorted_Radial)

        value = zip_data('AlongTrack', value)
        value = zip_data('CrossTrack', value)
        value = zip_data('Radial', value)

        if np.isnan(value['Forecast']['Target']['CrossTrack']['upper_bound']).all():
            value['Forecast']['Target']['CrossTrack']['upper_bound'] = []
            value['Forecast']['Target']['CrossTrack']['lower_bound'] = []
        if np.isnan(value['Forecast']['Chaser']['CrossTrack']['upper_bound']).all():
            value['Forecast']['Chaser']['CrossTrack']['upper_bound'] = []
            value['Forecast']['Chaser']['CrossTrack']['lower_bound'] = []
    
        value['Forecast'] = divide_by_model(value['Forecast'])
        #value['Forecast']['Chaser'] = subtract_days(value['Forecast']['Chaser'])
        #value['Forecast']['Target'] = subtract_days(value['Forecast']['Target'])

    new_datasets['HAC']['color'] = '#A1173C'
    new_datasets['ephemeris']['color'] = '#FFD521'

    return new_datasets
```
